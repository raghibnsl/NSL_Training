```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
x=np.random.randint(-20,20,100).reshape(-1,1)
y=x**3 + np.random.randint(1,3)
standardize=StandardScaler()
x_=standardize.fit_transform(x)
```

```python
plt.scatter(x_,y)
```

```python
l_reg=LinearRegression()
l_reg.fit(x_,y)

print(l_reg.intercept_,l_reg.coef_)
plt.scatter(x_,y)
plt.plot(np.linspace(-2,2,40).reshape(-1,1),l_reg.coef_*np.linspace(-2,2,40).reshape(-1,1)+l_reg.intercept_)
```

```python
x=np.random.randint(-20,20,100).reshape(-1,1)
y=x**3 + np.random.randint(1,20,100).reshape(-1,1)

poly_features=PolynomialFeatures(degree=3,include_bias=True)
x_poly=poly_features.fit_transform(x)
x_poly_=standardize.fit_transform(x_poly)
l_reg.fit(x_poly_,y)
plt.scatter(x,y)
def predict_(x,model,*features):
    for i in features:
        x=i.transform(x)
    return model.predict(x)
x_plot=np.linspace(x.min(),x.max(),200).reshape(-1,1)

plt.plot(x[np.argsort(x,axis=0)].reshape(-1,1),predict_(x[np.argsort(x,axis=0)].reshape(-1,1),l_reg,poly_features,standardize),c='r')
```

```python
l_reg.score(x_poly_,y)
```
