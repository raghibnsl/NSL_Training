```python
from sklearn.preprocessing import Normalizer
fig=plt.figure()
ax=fig.add_subplot(121,projection='3d')
ax1=fig.add_subplot(122,projection='3d')
df=pd.DataFrame({'x':np.random.randint(-80,80,1000),'y':np.random.randint(-100,100,1000),'z':np.random.randint(-150,150,1000)})
normalizer=Normalizer()
ax.scatter3D(df.x,df.y,df.z)
df_=normalizer.fit_transform(df)
ax1.scatter3D(df_[:,0],df_[:,1],df_[:,2])
```
